2. *Bedrooms and Price*: What is the relationship between the number of bedrooms and the price of the house? Does having more bedrooms increase the price?

```python
bedroom_price = df.groupby('bedrooms')['price'].mean()
plt.figure(figsize=(8, 5))
plt.plot(bedroom_price.index, bedroom_price.values, marker='o', linestyle='-', color='blue')
plt.title('Average House Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price ($)')
plt.xticks(bedroom_price.index)  # Ensures all bedroom counts are shown on the x-axis
plt.grid(True)
plt.show()
bedroom_price
```

9. *Renovation Impact*: Does the year of renovation (yr_renovated) affect the price of the house? Are recently renovated houses more expensive?

```python
df['yr_renovated'] = df['yr_renovated'].replace(0, 'Not Renovated')
renovation_price = df.groupby('yr_renovated')['price'].mean()
plt.figure(figsize=(10, 6))
renovation_price.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Average House Price by Year of Renovation')
plt.xlabel('Year of Renovation')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')
plt.show()
renovation_price
```

*Zip Code Price Comparison*: What are the average house prices for different zip codes? Which zip code has the highest and lowest average prices?

```python
zipcode_price = df.groupby('zipcode')['price'].mean()
highest_zipcode = zipcode_price.idxmax()
lowest_zipcode = zipcode_price.idxmin()
highest_price = zipcode_price.max()
lowest_price = zipcode_price.min()
plt.figure(figsize=(10, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(zipcode_price)))  # Generate a colormap
zipcode_price.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Average House Prices by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)  
plt.grid(axis='y')
plt.tight_layout()  
plt.show()
print("Average House Prices by Zip Code:")
print(zipcode_price)
print(f'\nZip code with the highest average price: {highest_zipcode} (${highest_price:.2f})')
print(f'Zip code with the lowest average price: {lowest_zipcode} (${lowest_price:.2f})')
```

*Condition and Price*: How does the condition of the house (as recorded in the condition column) affect its price?

```python
condition_price = df.groupby('condition')['price'].mean()
plt.figure(figsize=(10, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(condition_price)))  
condition_price.plot(kind='bar', color=colors, edgecolor='white')
plt.title('Average House Prices by Condition', fontsize=16, fontweight='bold')
plt.xlabel('Condition (1 = Poor, 5 = Excellent)', fontsize=14)
plt.ylabel('Average Price ($)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("Average House Prices by Condition:")
print(condition_price)
```

 *Age of the House*: How does the year the house was built (yr_built) impact the price? Are newer houses priced higher than older ones?

```python
year_price = df.groupby('yr_built')['price'].mean()
plt.figure(figsize=(10, 6))
colors = plt.cm.coolwarm(np.linspace(0, 1, len(year_price)))  # Generate a colormap
plt.plot(year_price.index, year_price.values, marker='o', linestyle='-', color='orange', markersize=8)
plt.fill_between(year_price.index, year_price.values, color='lightblue', alpha=0.3)
plt.title('Average House Prices by Year Built', fontsize=16, fontweight='bold')
plt.xlabel('Year Built', fontsize=14)
plt.ylabel('Average Price ($)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("Average House Prices by Year Built:")
print(year_price)
```

*Basement and Price*: Is there a significant price difference between houses with basements (sqft_basement) and those without?

```python
df['has_basement'] = df['sqft_basement'].apply(lambda x: 'With Basement' if x > 0 else 'Without Basement')
basement_price = df.groupby('has_basement')['price'].mean()
plt.figure(figsize=(10, 6))
colors = plt.cm.spring(np.linspace(0, 1, len(basement_price)))  # Generate a colormap
basement_price.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Average House Prices by Basement Presence', fontsize=16, fontweight='bold')
plt.xlabel('Basement Presence', fontsize=14)
plt.ylabel('Average Price ($)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("Average House Prices by Basement Presence:")
print(basement_price)
```

*Living Space Change*: Analyze how living space in 2015 (sqft_living15) compares to previous years (sqft_living) and how this change impacts the price.

```python
df['living_space_change'] = df['sqft_living15'] - df['sqft_living']
change_price = df.groupby('living_space_change')['price'].mean()
plt.figure(figsize=(10, 6))
colors = plt.cm.cool(np.linspace(0, 1, len(change_price)))  
change_price.plot(kind='bar', color=colors, edgecolor='b')
plt.title('Average House Prices by Change in Living Space', fontsize=16, fontweight='bold')
plt.xlabel('Change in Living Space (sqft)', fontsize=14)
plt.ylabel('Average Price ($)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("Average House Prices by Change in Living Space:")
print(change_price)
```

*Geographical Location and Price*: Use latitude (lat) and longitude (long) to explore how house prices vary across different locations. Are certain locations more expensive?

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['long'], df['lat'], c=df['price'], cmap='viridis', s=100, alpha=0.7, edgecolor='black')
cbar = plt.colorbar(scatter)
cbar.set_label('House Price ($)', fontsize=14)
plt.title('House Prices by Geographical Location', fontsize=16, fontweight='bold')
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Latitude', fontsize=14)
plt.tight_layout()
plt.show()
```
